````python
x = np.array([1, 2, 3, 4, 5, 6])
print("beginning x: \n", x)
x_reshape = x.reshape((2, 3))
print('reshaped x: \n', x_reshape)
````

numpy arrays are row major ordering which means they are specified as a sequence of rows

````python
# top left element
x_reshape[0, 0]
````

````python
# bottom right
x_reshape[1, 2]
````

````python
print('x before we modify x_reshape:\n', x)
print('x_reshape before we modify x_reshape:\n', x_reshape)
x_reshape[0, 0] = 5
print('x_reshape after we modify its top left element:\n', x_reshape)
print('x after we modify top left element of x_reshape:\n', x) # if we modify x_reshape x also gets modified
````

Modifying ```x_reshape``` also modified ```x``` because the two objects occupy the same space in the memory. 

we can modify arrays but we cannot modify tuples.

##### Mean, Variance and Standard Deviation

````python
rng = np.random.default_rng(3)
y = rng.standard_normal(10)
np.mean(y), y.mean()
````

````python
np.var(y), y.var()
````

````python
np.var?
````

##### Variance 
 


````python
np.mean((y- y.mean())**2), np.var(y)
````

##### Standard Deviation

$\sqrt{Variance}$

````python
np.sqrt(np.var(y)), np.std(y)
````

We can create more sophisticated plots like the contour plot that uses three-dimensional data. It takes 3 arguments:
- A vector of ```x``` values (the first dimension)
- A vector of ```y``` values (the second dimension)
- A matrix whose elements correspond to the ```z``` value (third dimension) for the pair of ```(x, y)``` coordinates.

``` np.linespace(a, b, n)``` returns a vector of ```n``` numbers starting at ```a```and ending at ```b```

````python
fig, ax = subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, f);
````

````python
fig, ax = subplots(figsize=(8, 8))
ax.contour(x, y, f, levels=45);
````

````python
fig, ax = subplots(figsize=(8, 8))
# heatmap vibes
ax.imshow(f);
````
